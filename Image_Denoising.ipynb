```python
IMG_WIDTH=200
IMG_HEIGHT=200
img_folder=r'D:\dataset'
```

```python
def create_dataset_original(original_folder):
   
    img_data_array1=[]
    class_name1=[]
   
    for dir1 in os.listdir(original_folder):
        for file in os.listdir(os.path.join(original_folder, dir1)):
       
            image_path= os.path.join(original_folder, dir1,  file)
            image= cv2.imread( image_path, cv2.COLOR_BGR2RGB)
            image=cv2.resize(image, (IMG_HEIGHT, IMG_WIDTH),interpolation = cv2.INTER_AREA)
            image=np.array(image)
            image = image.astype('float32')
            image /= 255 
            img_data_array1.append(image)
            class_name1.append(dir1)
    return img_data_array1, class_name1
# extract the image array and class name
```

```python
img_data, class_name1 =create_dataset_original(r'D:\dataset')
```
